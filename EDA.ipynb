# Step 1: Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 2: Load the Dataset

```python
df = pd.read_csv('train.csv') 
```

# Step 4: Check for Missing Values

```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.head(2)
```

# Step 5: Univariate Analysis

### Categorical Features

```python
sns.set_style("whitegrid")
plt.figure(figsize=(5,5))
sns.countplot(x='Survived', data=df,color='blue')
plt.title("Survival Distribution",fontsize=16)
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

### Numerical Features

```python
sns.histplot(df['Age'].dropna(), bins=30, kde=True,color='blue')
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x='Fare', data=df)
plt.title("Fare Distribution")
plt.show()
```

# Step 6: Bivariate Analysis

### Survival by Sex

```python
sns.countplot(x='Survived', hue='Sex', data=df)  # 0=no servive
plt.title("Survival by Gender")                  # 1= yea servive
plt.show()
```

### Survival by Class 

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

### Age vs Survival 

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

#  Step 7: Correlation Heatmap


```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Step 8: Pairplot 


```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']].dropna(), hue='Survived')
plt.show()
```
